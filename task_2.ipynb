```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.style.use("ggplot")
```

```python
df = pd.read_csv("titanic_cleaned.csv")
df.head()
```

```python
print(df.describe())
```

```python
print(df['Sex'].value_counts())
print(df['Embarked_Q'].value_counts())
print(df['Embarked_S'].value_counts)
```

```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1) 
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col]) 
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# the correlation matrix
correlation_matrix = df[numerical_cols].corr() 

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='viridis') # Adjust columns
plt.suptitle('Pairplot of Selected Features by Survival', y=1.02) # Adjust title position
plt.show()
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Age Distribution by Survival')
plt.show()
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival Count by Passenger Class')
plt.show()
```
